```python
WEAVIATE_URL = ""
WEAVIATE_API_KEY = ""
OPENAI_API_KEY = ""

import weaviate

# Note, we are using OpenAI API to generate the vector embeddings. You can use any other vector embedding as well.

client = weaviate.Client(
    url=WEAVIATE_URL,
    auth_client_secret=weaviate.AuthApiKey(api_key=WEAVIATE_API_KEY),
    additional_headers={
        "X-OpenAI-Api-Key": OPENAI_API_KEY
    })

# Test connection
print(client.is_live())
print(client.is_ready())
```

```python
class_obj = {
    "class": "Product_vectors",
    "description": "Product in the products catalog",
    "vectorIndexType": "hnsw",
    "vectorIndexConfig": {
        "distance": "cosine", # Similarity metric
        "efConstruction": 128, # Default value - trades off index build speed against index search speed
        "maxConnections": 64 # Default value - max number of connections supported
    },
    "vectorizer": "text2vec-openai",
    "properties": [
        {
            "dataType": [
                "int"
            ],
            "description": "ID of the product",
            "name": "product_id",
            "indexInverted": True,
            "moduleConfig": {
                "text2vec-openai": {
                  "skip": True
                }
              }
        },
        {
            "dataType": [
                "text"
            ],
            "description": "Name of the product",
            "name": "name",
            "indexInverted": True,
            "moduleConfig": {
                "text2vec-openai": {
                  "skip": False,
                  "vectorizePropertyName": False
                }
              }
        },
        {
            "dataType": [
                "text"
            ],
            "description": "Description of the product",
            "name": "description",
            "indexInverted": True,
            "moduleConfig": {
                "text2vec-openai": {
                  "skip": False,
                  "vectorizePropertyName": False
                }
            }
        }
    ]
}

# If you need to delete existing class
client.schema.delete_class("Product_vectors")

client.schema.create_class(class_obj)
```
